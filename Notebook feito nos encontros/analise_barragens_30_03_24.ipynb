```python
barragem['Teor (%) do minério principal inserido no rejeito'] = barragem['Teor (%) do minério principal inserido no rejeito'].str.replace(
    'Não se aplica a esse tipo de barragem','0'
)
barragem['Teor (%) do minério principal inserido no rejeito'] = barragem['Teor (%) do minério principal inserido no rejeito'].str.replace(',','.')
barragem['Teor (%) do minério principal inserido no rejeito'] = barragem['Teor (%) do minério principal inserido no rejeito'].astype(float)
```

```python
sns.histplot(x='Teor (%) do minério principal inserido no rejeito',data=barragem)
```

```python
intervalo_por = []
for i in list(barragem['Teor (%) do minério principal inserido no rejeito']):
  if i <= 20:
    intervalo_por.append(1)
  elif i > 20 and i <=40:
    intervalo_por.append(2)
  elif i>40 and i<=60:
    intervalo_por.append(3)
  else:
    intervalo_por.append(4)
```

```python
barragem['teor (%) tipo'] = intervalo_por
barragem.head()
```

```python
sns.countplot(x='teor (%) tipo',data=barragem)
```

```python
sns.countplot(x='UF',data=barragem,hue='teor (%) tipo',palette='pastel')
```

```python
sns.countplot(y='Município',data=barragem[barragem['UF']=='PA']
              ,hue='teor (%) tipo',palette='pastel')
```

```python
sns.histplot(y='Tipo de barragem quanto ao material de construção',data=barragem)
```

```python
sns.countplot(y='Tipo de barragem quanto ao material de construção',data=barragem,hue='teor (%) tipo',
              palette='pastel')
```
