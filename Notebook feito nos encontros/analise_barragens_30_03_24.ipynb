```python
barragem['Teor (%) do minério principal inserido no rejeito'] = barragem['Teor (%) do minério principal inserido no rejeito'].str.replace(
    'Não se aplica a esse tipo de barragem','0'
)
barragem['Teor (%) do minério principal inserido no rejeito'] = barragem['Teor (%) do minério principal inserido no rejeito'].str.replace(',','.')
barragem['Teor (%) do minério principal inserido no rejeito'] = barragem['Teor (%) do minério principal inserido no rejeito'].astype(float)
```

```python
plt.title('Teor (%) do teor do minério principal inserido no rejeito',
          fontsize=14)
sns.histplot(x='Teor (%) do minério principal inserido no rejeito',data=barragem)
plt.xlabel('Porcentagem')
plt.ylabel('Quantidade');
```

```python
plt.figure(figsize=(12,6))
plt.title('Teor (%) do teor do minério principal inserido no rejeito para valores maior que zero',
          fontsize=14)
sns.histplot(x='Teor (%) do minério principal inserido no rejeito',
             data=barragem[barragem['Teor (%) do minério principal inserido no rejeito']>0],
             kde=True)
plt.xlabel('Porcentagem')
plt.ylabel('Quantidade');
```

```python
intervalo_por = []
for i in list(barragem['Teor (%) do minério principal inserido no rejeito']):
  if i == 0:
    intervalo_por.append(1)
  elif i> 0 and i <=20:
    intervalo_por.append(2)
  elif i > 20 and i <=40:
    intervalo_por.append(3)
  elif i>40 and i<=60:
    intervalo_por.append(4)
  else:
    intervalo_por.append(5)
```

```python
barragem['teor (%) tipo'] = intervalo_por
barragem.head()
```

```python
plt.title('Teor (%) de minerio de barragem de acordo com a classe',
          fontsize=14)
sns.countplot(x='teor (%) tipo',data=barragem)
plt.xlabel('Classe')
plt.ylabel('Quantidade')
```

```python
plt.figure(figsize=(12,6))
plt.title('Teor de minerio de acordo com o Estado',
          fontsize=14)
sns.countplot(x='UF',data=barragem,hue='teor (%) tipo',palette='tab10')
plt.xlabel('Estado')
plt.ylabel('Quantidade')
plt.legend(title='Teor (%) por classse')
```

```python
sns.countplot(y='Município',data=barragem[barragem['UF']=='PA']
              ,hue='teor (%) tipo',palette='pastel')
```

```python
plt.title('Tipo de barragem quanto ao material de construção',
          fontsize=14)
sns.histplot(y='Tipo de barragem quanto ao material de construção',data=barragem)
plt.xlabel('Quantidade')
plt.ylabel('Tipo de construção')
```

```python
plt.figure(figsize=(12,6))
plt.title('Tipo de barragem quanto ao material de construção em relação ao teor da barragem',
          fontsize=14)
sns.countplot(y='Tipo de barragem quanto ao material de construção',data=barragem,hue='teor (%) tipo',
              palette='tab10')
plt.xlabel('Quantidade')
plt.ylabel('Tipo de ematerial')
plt.legend(title='Teor (%) por classse')
```
