```python
import numpy as np
import pandas as pd

import re
import string
from nltk.corpus import stopwords
from nltk import word_tokenize
import nltk

from wordcloud import WordCloud, STOPWORDS, ImageColorGenerator
import matplotlib.pyplot as plt
```

```python
noticias = pd.read_csv('titulos tirado do google news.csv')
noticias.drop('Unnamed: 0',axis=1,inplace=True)
noticias.head()
```

```python
noticias['Titulo da noticia'] = noticias['Titulo da noticia'].str.lower()
noticias['Descrição da noticia 	'] = noticias['Descrição da noticia'].str.lower()
```

```python
noticias['Titulo da noticia'] = noticias['Titulo da noticia'].str.replace('em','')
noticias['Titulo da noticia'] = noticias['Titulo da noticia'].str.replace('de','')
```

```python
noticias['Titulo da noticia'] = noticias['Titulo da noticia'].str.replace('para','')
noticias['Titulo da noticia'] = noticias['Titulo da noticia'].str.replace('da','')
```

```python
noticias.head()
```

```python
titulo = " ".join(i for i in noticias['Titulo da noticia'])
stopwords = set(STOPWORDS)
wordcloud = WordCloud(stopwords=stopwords,
                      background_color="black").generate(titulo)
plt.figure( figsize=(12,6))
plt.imshow(wordcloud, interpolation='bilinear')
plt.axis("off")
plt.show()
```

As principais termos que aparecem é em relação a vale, minas as cidede de ouro preto e mariana e forquilha

```python
noticias['Descrição da noticia'] = noticias['Descrição da noticia'].str.replace('de','')
noticias['Descrição da noticia'] = noticias['Descrição da noticia'].str.replace('da','')
noticias['Descrição da noticia'] = noticias['Descrição da noticia'].str.replace('que','')
noticias['Descrição da noticia'] = noticias['Descrição da noticia'].str.replace('em','')
noticias['Descrição da noticia'] = noticias['Descrição da noticia'].str.replace('um','')
noticias['Descrição da noticia'] = noticias['Descrição da noticia'].str.replace('uma','')
```

```python
noticias['Descrição da noticia'] = noticias['Descrição da noticia'].str.replace('na','')
```
