```python
import pandas as pd
import datetime
import requests
from xml.etree import ElementTree
```

```python

def pesquisar_noticias(string):

    titulos = []
    datas_publicacao = []
    links = []
    canais = []

    print('\n>>>>>>>>>>>>>>>>> Extraindo notícias para a string {0}'.format(string))
    url = 'https://news.google.com/rss/search?q='
    parametros = '&hl=pt-BR&gl=BR&ceid=BR%3Apt-419'
    url_final = url + string + parametros
    response = requests.get(url_final)

    if response.status_code == 200:

        noticias = ElementTree.fromstring(response.content).find('channel')
        noticias = list(noticias)

        noticias = noticias[8:]

        print('Foram encontrados {0} resultados'.format(len(noticias)))

        for noticia in noticias:
            titulo = noticia.findtext('title')
            data_publicacao = noticia.findtext('pubDate')
            link = noticia.findtext('link')
            canal = noticia.findtext('source')

            titulos.append(titulo)
            datas_publicacao.append(data_publicacao)
            links.append(link)
            canais.append(canal)
```

```python
pesquisar_noticias('Barragem de mineração')
```

```python
dados_noticias = {
    "Titulo": titulos,
    "Data de Publicação": datas_publicacao,
    "Link": links,
    "Canal": canais
}

dataframe_noticias = pd.DataFrame(dados_noticias)
```

```python
termo_utilizado = []

for i in range(dataframe_noticias.shape[0]):
  termo_utilizado.append('Barragem de mineração')
```

```python
dataframe_noticias['Termo utilizado'] = termo_utilizado
dataframe_noticias.head()
```
