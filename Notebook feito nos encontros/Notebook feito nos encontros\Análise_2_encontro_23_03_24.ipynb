```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
barragem = pd.read_csv('/content/drive/MyDrive/Datasets/Barragens.csv',encoding='latin1')
barragem.drop('ID',axis=1,inplace=True)
barragem.head()
```

```python
barragem['Capacidade Total do Reservatório (m³)'] = barragem['Capacidade Total do Reservatório (m³)'].str.replace(',','.')
barragem['Capacidade Total do Reservatório (m³)'] = barragem['Capacidade Total do Reservatório (m³)'].astype(float)
barragem.head()
```

```python
sns.boxplot(x='Capacidade Total do Reservatório (m³)',data=barragem,showfliers=False)
```

```python
plt.figure(figsize=(12,6))
sns.stripplot(x='UF',y='Capacidade Total do Reservatório (m³)',data=barragem,hue='Categoria de Risco - CRI')
```
