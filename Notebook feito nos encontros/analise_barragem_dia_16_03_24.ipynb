# Dados sobre a barragem



```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
barragem = pd.read_csv('/content/drive/MyDrive/Datasets/Barragens.csv',encoding='latin-1')
barragem.drop('ID',axis=1,inplace=True)
barragem.head()
```

```python
barragem.shape
```

```python
plt.figure(figsize=(12,6))
plt.title('Barragens por Estado',fontsize=14)
sns.histplot(x='UF',data=barragem,color='royalblue',shrink=.8)
plt.xlabel('Estado')
plt.ylabel('Quantidade')
```

```python
barragem['UF'].value_counts(normalize=True)
```

```python
plt.title('Municipios do Estado do Pará onde tem barragem',fontsize=14)
sns.histplot(y='Município',data=barragem[barragem['UF']=='PA'],color='royalblue')
plt.xlabel('Quantidade')
```

```python
plt.figure(figsize=(12,10))
plt.title('Municipios do Estado de Minas Gerais onde tem barragem',fontsize=14)
sns.histplot(y='Município',data=barragem[barragem['UF']=='MG'],color='royalblue')
plt.xlabel('Quantidade')
```

```python
plt.title('Categoria de risco das barragens',fontsize=14)
sns.histplot(x='Categoria de Risco - CRI',data=barragem,color='royalblue',
             shrink=.8)
plt.xlabel('Categoria')
plt.ylabel('Quantidade')
```

```python
barragem['Categoria de Risco - CRI'].value_counts(normalize=True)
```

```python
plt.title('Dano de potencial associado a barragem',
          fontsize=14)
sns.histplot(x='Dano Potencial Associado - DPA',data=barragem,color='royalblue',
             shrink=.8)
plt.xlabel('Categroia')
plt.ylabel('Quantidade')
```

```python
barragem['Dano Potencial Associado - DPA'].value_counts(normalize=True)
```

```python
plt.figure(figsize=(12,6))
plt.title('Estados onde está localizado as barragens de acordo com o risco',
          fontsize=14)
sns.histplot(x='UF',data=barragem,color='royalblue',hue='Categoria de Risco - CRI',shrink=.8,multiple='dodge')
plt.xlabel('Estado')
plt.ylabel('Quantidade')
```

```python
plt.figure(figsize=(12,6))
plt.title('Estados onde está localizado as barragens de acordo com o dano potencial',
          fontsize=14)
sns.histplot(x='UF',data=barragem,color='royalblue',hue='Dano Potencial Associado - DPA',shrink=.8,multiple='dodge')
plt.xlabel('Estado')
plt.ylabel('Quantidade')
```
