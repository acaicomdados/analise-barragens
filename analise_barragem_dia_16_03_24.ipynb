# Dados sobre a barragem



```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
barragem = pd.read_csv('/content/drive/MyDrive/Datasets/Barragens.csv',encoding='latin-1')
barragem.drop('ID',axis=1,inplace=True)
barragem.head()
```

```python
barragem.shape
```

```python
plt.figure(figsize=(12,6))
plt.title('Barragens por Estado',fontsize=14)
sns.histplot(x='UF',data=barragem,color='royalblue')
plt.xlabel('Estado')
plt.ylabel('Quantidade')
```

```python
plt.title('Municipios do Estado do Pará onde tem barragem',fontsize=14)
sns.histplot(y='Município',data=barragem[barragem['UF']=='PA'],color='royalblue')
plt.xlabel('Quantidade')
```

```python
plt.figure(figsize=(12,10))
plt.title('Municipios do Estado de Minas Gerais onde tem barragem',fontsize=14)
sns.histplot(y='Município',data=barragem[barragem['UF']=='MG'],color='royalblue')
plt.xlabel('Quantidade')
```
